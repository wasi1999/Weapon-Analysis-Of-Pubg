```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pubg=pd.read_csv('Pubg Dataset.csv')
pubg.head()
```

```python
plt.figure(figsize=(16,7))
sns.countplot(pubg['Weapon Type'],palette='viridis')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(pubg['Bullet Type'],palette='YlOrBr')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(pubg['Magazine Capacity'],palette='rocket_r')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(pubg['Weapon Type'],pubg['Damage'],palette='YlOrBr')
```

```python
most_damaging_weapon=pubg[pubg['Weapon Type']=='Assault Rifle'][['Weapon Name','Damage']]
most_damaging_weapon
```

```python
plt.figure(figsize=(10,5))
sns.barplot(most_damaging_weapon['Weapon Name'],most_damaging_weapon['Damage'],palette='icefire')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(pubg['Bullet Type'],pubg['Range'],palette='rocket')
```

```python
pubg[pubg['Weapon Type']=='Assault Rifle'][['Weapon Name','Damage']]sns.barplot(pubg['Weapon Type'],pubg['Range'],palette='cubehelix')
```

```python
plt.figure(figsize=(10,6))
Rifle_range=pubg[pubg['Weapon Type']=='Pistol'][['Weapon Name','Range']]
sns.barplot(Rifle_range['Weapon Name'],Rifle_range['Range'],palette='RdPu')
```
